```python
import pickle
import shap

with open("shap_values_correct_100.pkl", "rb") as f:
    obj = pickle.load(f)

shap_values = obj["shap_values"]
meta = obj["data"]
label_order = obj["meta"]["label_order"]

i = 40
cls = label_order.index(meta[i]["pred"])

shap.plots.text(shap_values[i, :, cls])
```

```python
import numpy as np
# 输出高置信度样本
# 假设 shap_values.base_values 可用
confidence = shap_values.base_values.max(axis=1)

top_idx = np.argsort(-confidence)[:5]
print(top_idx)
```
